```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# === Функции для генерации данных ===
def create_sample(dist_func, n, dist_params):
    return dist_func(*dist_params, size=n)

# === Функции анализа данных ===
def get_quantiles(sample, q_list):
    return {f"q{q}": np.percentile(sample, q) for q in q_list}

def calculate_confidence_interval(sample, confidence_level):
    #Рассчитывает границы доверительного интервала для среднего значения выборки.
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    z = norm.ppf(1 - (1 - confidence_level) / 2)
    margin_error = z * sample_std / np.sqrt(len(sample))
    return sample_mean - margin_error, sample_mean + margin_error

def required_sample_size(confidence_level, error_margin, std_dev):
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    return int(np.ceil((z_score * std_dev / error_margin) ** 2))

# === Визуализация данных ===
def plot_sample_with_ci(sample, ci_bounds, graph_title):
    """
    Отображает выборку и её доверительные интервалы на гистограмме.

    :param sample: Данные выборки
    :param ci_bounds: Границы доверительного интервала
    :param graph_title: Название графика
    """
    plt.figure(figsize=(8, 5))
    plt.hist(sample, bins=25, color='lightblue', edgecolor='black', alpha=0.7)
    plt.axvline(ci_bounds[0], color='red', linestyle='--', label=f'Lower CI: {ci_bounds[0]:.2f}')
    plt.axvline(ci_bounds[1], color='green', linestyle='--', label=f'Upper CI: {ci_bounds[1]:.2f}')
    plt.title(graph_title)
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.legend()
    plt.show()

# === Основной блок анализа ===
confidence_levels = [0.90, 0.95, 0.99]
error_margins = [0.1, 0.05, 0.01]

# Распределения для анализа
distributions_info = {
    'Uniform': (np.random.uniform, (0, 1)),
    'Normal': (np.random.normal, (0, 1)),
    'Cauchy': (np.random.standard_cauchy, ())
}

# Анализ каждого распределения
for dist_name, (dist_func, dist_params) in distributions_info.items():
    print(f"\Анализ распределения: {dist_name}")

    for conf_lvl in confidence_levels:
        for err_margin in error_margins:
            # Рассчёт минимального размера выборки
            sample_size = required_sample_size(conf_lvl, err_margin, std_dev=1)
            sample_data = create_sample(dist_func, sample_size, dist_params)

            # Квантили и доверительные интервалы
            quantiles = get_quantiles(sample_data, [1, 95])
            ci_lower, ci_upper = calculate_confidence_interval(sample_data, conf_lvl)

            # Проверка попадания доверительных интервалов в диапазон выборки
            is_within_sample = ci_lower >= sample_data.min() and ci_upper <= sample_data.max()

            # Вывод результатов
            print(f"\nУровень доверия: {conf_lvl}, Погрешность: {err_margin}")
            print(f"Размер выборки: {sample_size}")
            print(f"1%-квантиль: {quantiles['q1']:.3f}, 95%-квантиль: {quantiles['q95']:.3f}")
            print(f"Доверительный интервал: ({ci_lower:.3f}, {ci_upper:.3f}), В пределах выборки: {is_within_sample}")

            # Построение графика
            plot_sample_with_ci(sample_data, (ci_lower, ci_upper),
                                graph_title=f'{dist_name} Distribution (Conf: {conf_lvl}, Error: {err_margin})')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def create_sample(distribution, params, size):
    return distribution(*params, size=size)

# Оценка погрешности методом Монте-Карло 
def monte_carlo_uncertainty(sample, iterations=10000):
    means = np.array([
        np.mean(np.random.choice(sample, size=len(sample), replace=True))
        for _ in range(iterations)
    ])
    return np.std(means)

# Метод линеаризации
def linearization_uncertainty(sample):

    mean = np.mean(sample)
    std_dev = np.std(sample, ddof=1)

    mean_error = std_dev / np.sqrt(len(sample))
    variance_error = (2 * np.var(sample, ddof=1)**2 / (len(sample) - 1))**0.5

    return mean_error, variance_error

# Метод Крейновича
def kreinovich_uncertainty(sample):
    n = len(sample)
    std_dev = np.std(sample, ddof=1)

    mean_error = std_dev / np.sqrt(n)
    variance_error = (2 * np.var(sample, ddof=1)**2) / (n - 1)

    return mean_error, variance_error

# Визуализация погрешностей
def compare_uncertainties(errors, methods):
    categories = ['Среднее', 'Дисперсия']
    x = np.arange(len(categories))

    fig, ax = plt.subplots(figsize=(8, 5))
    width = 0.25

    for i, (method, error) in enumerate(zip(methods, errors)):
        ax.bar(x + i * width, error, width, label=method)

    ax.set_xticks(x + width)
    ax.set_xticklabels(categories)
    ax.set_ylabel('Погрешность')
    ax.set_title('Сравнение методов оценки погрешностей')
    ax.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

distribution = np.random.normal
params = (0, 1)
sample_size = 100

# Генерация выборки
data_sample = create_sample(distribution, params, sample_size)

# Вычисление погрешностей
mc_error = monte_carlo_uncertainty(data_sample)
lin_mean_error, lin_variance_error = linearization_uncertainty(data_sample)
krein_mean_error, krein_variance_error = kreinovich_uncertainty(data_sample)

# Вывод результатов
print(f"Метод Монте-Карло: Погрешность = {mc_error:.4f}")
print(f"Метод линеаризации: Среднее = {lin_mean_error:.4f}, Дисперсия = {lin_variance_error:.4f}")
print(f"Метод Крейновича: Среднее = {krein_mean_error:.4f}, Дисперсия = {krein_variance_error:.4f}")

# Сравнение методов
compare_uncertainties(
    errors=[
        [mc_error, mc_error],
        [lin_mean_error, lin_variance_error],
        [krein_mean_error, krein_variance_error]
    ],
    methods=['Монте-Карло', 'Линеаризация', 'Крейнович']
)
```
